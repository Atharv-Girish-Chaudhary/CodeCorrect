```python
# Sample Input 2
s1 = "pritn"
s2 = "print"
print("s1-",s1)
print("s2-",s2)
```

```python
# Assigning the length of the strings
m = len(s1)
n = len(s2)
print("Lenth of s1-",m)
print("Lenth of s2-",n)
```

```python
def print_dp_table():
    
    # Column headers
    print("     ", end="")
    print(f"{'ε':>4}", end="")
    for ch in s2:
        print(f"{ch:>4}", end="")
    print()
    
    # Separator
    print("    " + "----" * (n + 1))
    
    # Rows
    for i in range(m + 1):
        label = "ε" if i == 0 else s1[i - 1]
        print(f"  {label} |", end="")
        for j in range(n + 1):
            print(f"{dp[i][j]:>4}", end="")
        print()
```

```python
# Creating the table
dp = [[0] * (n + 1) for _ in range(m + 1)]
print("Table-\n")
print_dp_table()
```

```python
# Base Case
for i in range(m + 1):
        dp[i][0] = i  
for j in range(n + 1):
        dp[0][j] = j  
print("Base Case -\n")
print_dp_table()
```

```python
# Filling the table using Bottom-up
for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] 
            else:
                dp[i][j] = 1+min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1])  
print("Final Table-\n")
print_dp_table()
```
