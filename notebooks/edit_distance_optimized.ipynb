```python
# Sample Input 1
s1 = "cat"
s2 = "car"
print("s1-",s1)
print("s2-",s2)
```

```python
# Sample Input 2
s1 = "kitten" 
s2 = "sitting" 
print("s1-",s1)
print("s2-",s2)
```

```python
# Sample Input 2
s1 = "pritn"
s2 = "print"
print("s1-",s1)
print("s2-",s2)
```

```python
if len(s1) < len(s2):
    s1, s2 = s2, s1
```

```python
# Assigning the length of the strings
m, n = len(s1), len(s2)
print("Lenth of s1-",m)
print("Lenth of s2-",n)
```

```python
# Creating the rows
prev = list(range(n + 1))  # base case: dp[0][j] = j
curr = [0] * (n + 1)
print("Previous row",prev)
print("Current row",curr)
```

```python
print(f"Row {0}\t(''):prev={prev}-> updated curr={curr}")
for i in range(1, m + 1):
        curr[0] = i
        print(f"Row {i} ('{s1[i-1]}'): prev={prev}\n\t  -> curr={curr}",end="")
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                curr[j] = prev[j - 1]
            else:
                curr[j] = 1 + min(
                    prev[j],
                    curr[j - 1],
                    prev[j - 1],
                )
        print(f"-> updated curr={curr}")
        prev, curr = curr, prev
```

```python
# Printing the final result for no of operations needed 
print(prev[n])
```
